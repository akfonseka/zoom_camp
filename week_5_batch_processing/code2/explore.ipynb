```python
import pandas as pd 
import pyarrow 
import os 
```

```python
csv_name = 'fhvhv_tripdata_2021-01.csv'
parquet_name = 'fhvhv_tripdata_2021-01.parquet'
# download the csv 
url = 'https://d37ci6vzurychx.cloudfront.net/trip-data/fhvhv_tripdata_2021-01.parquet'
os.system(f"wget {url} -O {parquet_name}")

# Read the Parquet file into a DataFrame
parquet_file = rf'{parquet_name}'
df_parq = pd.read_parquet(parquet_file, engine= 'pyarrow')

# Convert the DataFrame to CSV
df_parq.to_csv(csv_name, index=False)
```
